```octave
angles=linspace(0,2*pi,100);
y=sin(angles);
plot(angles, y,'ro');
grid on;
xlabel('Angle');
ylabel('Value');
text (pi, 0.7, "arbitrary text"); 
legend ("sin (x)")
#title('Graph'); # not working for some reason
```

```octave
# multiple graphs
plot(angles,y,':',angles,cos(angles),'-');
legend('Sine', 'Cosine');
```

```octave
plot(angles, tan(angles));
```

```octave
% Script to calculate and plot a rectified sine wave
t = linspace(0, 10, 100);
y = abs(sin(t)); %The abs command makes all negative numbers positive
plot(t,y);
title('Rectified Sine Wave');
xlabel('t');
```
