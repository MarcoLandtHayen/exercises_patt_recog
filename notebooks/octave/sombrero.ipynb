### Setup

Used [tutorial](https://thicolares.com/2018/11/11/creating-a-docker-image-for-jupyter-Octave.html) to build Docker Container connecting Jupyter notebooks with Octave, including gnuplot.

```octave
tx = ty = linspace (-8, 8, 41)';
[xx, yy] = meshgrid (tx, ty);
r = sqrt (xx .^2 + yy .^2) + eps;
tz = sin (r) ./ r;
mesh (tx, ty, tz);
xlabel ("tx");
ylabel ("ty");
zlabel ("tz");
title ("3-D Sombrero plot");
```
